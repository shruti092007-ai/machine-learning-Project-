```python
#LogisticRegression Example 1
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

X = np.array([1,2,3,4,5]).reshape(-1,1)
#result(0=Fail, 1=pass)
y = np.array([0, 0, 1, 1, 1])

#Create model
model = LogisticRegression()
model.fit(X,y)
#Create smooth curve
X_test = np.linspace(0, 6, 100).reshape(-1, 1)
y_prob= model.predict_proba(X_test)[:, 1]


#plot data points
plt.scatter(X,y)
#plot logistics curve
plt.plot(X_test, y_prob)

#Labels
plt.xlabel("Study Hours")
plt.ylabel("Probability of Passing")
plt.title("Logistics Regression:Pass/Fail Prediction")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = np.array([
    [22,2000],
    [25,2200],
    [23,2100],
    [45,8000],
    [50,8500],
    [48,8200]])

model = KMeans(n_clusters=2, n_init='auto')
model.fit(x)
labels = model.labels_
centers = model.cluster_centers_

# Fixed the loop syntax and indentation
for i in range(len(x)):
    if labels[i] == 0:
        plt.scatter(x[i][0], x[i][1], color='blue')
    else:
        plt.scatter(x[i][0], x[i][1], color='red')

# Corrected indentation and variable name (Centers -> centers)
plt.scatter(centers[:, 0], centers[:, 1], color='green', marker='x', s=200, label='Centers')
plt.xlabel("Age")
plt.ylabel("Monthly Spending")
plt.title("Customer Clustering using K-Means")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.array([
    [40,45],
    [42,50],
    [38,43],
    [85,88],
    [90,92],
    [87,85],
    [60,65],
    [62,63]])
inertia = []
for k in range(1, 6):
    Kmeans = KMeans(n_clusters=k, random_state=42)
    Kmeans.fit(X)
    inertia.append(Kmeans.inertia_)
plt.figure()
plt.plot(range(1, 6), inertia, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.array([
    [40,45],
    [42,50],
    [38,43],
    [85,88],
    [90,92],
    [87,85],
    [60,65],
    [62,63]])
inertia = []
for k in range(1, 6):
    Kmeans = KMeans(n_clusters=k, random_state=42)
    Kmeans.fit(X)
    inertia.append(Kmeans.inertia_)
plt.figure()
plt.plot(range(1, 6), inertia, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()

# GRAPH 2: Cluster Visualization
# ===========================
model = KMeans(n_clusters=3, random_state=42)
model.fit(X) # Fixed: changed x to X

labels = model.labels_ # Fixed: changed labets to labels
centers = model.cluster_centers_

colors = ['blue', 'red', 'green']

plt.figure()
for i in range(len(X)):
    # Now labels corresponds to the current model fit on X
    plt.scatter(X[i][0], X[i][1], color=colors[labels[i]])

plt.scatter(centers[:, 0], centers[:, 1], color='black', marker='x', s=200, label='Centers')

plt.xlabel("Math Marks")
plt.ylabel("Science Marks")
plt.title("Student Clustering")
plt.legend()
plt.show()

# ========================
#GRAPH 3 :Cluster Size Bar Chart
#==========================

unique_labels, counts = np.unique(labels, return_counts=True)

plt.figure()
plt.bar(unique_labels, counts)
plt.xlabel("Cluster Number")
plt.ylabel("Number of Students")
plt.title("Cluster Size Comparison")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
X = np.array([
    [40,45,42],
    [42,50,48],
    [38,43,40],
    [85,88,90],
    [90,92,95],
    [87,85,89],
    [60,65,63],
    [62,63,60]])

#Train Model
model = KMeans(n_clusters=3, random_state=42)
model.fit(X)
labels = model.labels_
centers = model.cluster_centers_

#Heatmap of Marks
plt.figure()
sns.heatmap(X, annot=True, cmap="coolwarm",xticklabels=["Math","Science","English"])
plt.title("Student Marks Heatmap")
plt.show()

#3D Clustering
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for i in range(len(X)):
    ax.scatter(X[i][0], X[i][1], X[i][2],color=colors[labels[i]])

ax.scatter(centers[:,0],centers[:,1],centers[:,2],color='black',marker='x',s=200)

ax.set_xlabel("Math")
ax.set_ylabel("Science")
ax.set_zlabel("English")
plt.title("3D Student Clustering")
plt.show()

```
